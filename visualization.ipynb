```python
import pandas as pd
import numpy as np

df = pd.read_csv("datasets/dataset_train.csv")
df.head()
```

```python
#Normalisation
def get_numeric_col_names(df):
    return df.select_dtypes(include=np.number).columns.tolist()

courses = get_numeric_col_names(df)
courses.remove('Index')

normalized_df = df.copy()
# for course in courses:
#     normalized_df[course] = (normalized_df[course]-normalized_df[course].mean())/normalized_df[course].std()
```

```python
import matplotlib.pyplot as plt

for course in courses:
    plt.figure()
    normalized_df.groupby(['Hogwarts House'])[course].plot(kind='hist', alpha=.5)
    plt.legend()
    plt.title(course)
```
